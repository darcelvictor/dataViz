# Notes en fonction des cours particuliers

```python
a = sb.countplot(x="G3", hue="paid", data=df).set_title('Notes en fonction des cours particuliers')
```

```python
print("Taux de Réussite chez les élèves qui prennent des cours ",len(df[ (df['G3']>=10) & (df['paid']=='yes')])/len(df['G3']))
print("Taux de Réussite chez les élèves qui ne prennent pas des cours ",len(df[ (df['G3']>=10) & (df['paid']=='no')])/len(df['G3']))
```

```python
a = sb.countplot(x="G3", hue="guardian", data=df).set_title('Notes en fonction de la personne avec qui l étudiant vit')
```

```python
print(len(df[ (df['G3']>=10) & (df['guardian']=='mother')])/len(df['G3']))
print(len(df[ (df['G3']<10) & (df['guardian']=='mother')])/len(df['G3']))

print(len(df[ (df['G3']>=10) & ((df['guardian']=='father')|(df['guardian']=='other'))])/len(df['G3']))
print(len(df[ (df['G3']<10) & ((df['guardian']=='father')|(df['guardian']=='other'))])/len(df['G3']))
```
